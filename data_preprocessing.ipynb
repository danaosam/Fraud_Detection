```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

```

```python
frd_df = pd.read_csv('fraud.csv')
```

```python
frd_df.info()
```

```python
frd_df.head()
```

```python
frd_df.columns
```

```python
missing_values_count = frd_df.isnull().sum()
missing_values_count
```

```python
# One-hot encoding the 'age' and 'gender' categorical columns
encoded_data = pd.get_dummies(frd_df, columns=['age', 'gender','category','merchant'])

encoded_data.head()
```

```python
encoded_data.columns
```

```python
#Normalize the numerical variable 
scaler = MinMaxScaler()
encoded_data['amount_normalized'] = scaler.fit_transform(encoded_data[['amount']])
```

```python
encoded_data.to_csv('preprocessed_df.csv')
```
